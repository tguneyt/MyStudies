# 5. If ....elif ..... else

![title](if_body.png)

```python
apple = 33
banana = 200
if banana < apple:
    print("banana is more expensive than apple")
```

```python
apple = 3000
banana = 3000

if banana > apple:
    print("banana is more expensive than apple")
elif banana < apple:
    print("banana is cheaper than apple")
else:
    print("banana is equal to apple")
```

```python
sira = input("Yarismadaki siraniz : ")

if sira == "1":
    print("Tebrikler. Kazandiniz")
else:
    print("Cok uzgunuz. Kaybettiniz")
```

# 6. while .... for 

While Sart saglanincaya kadar devam etmeyi saglar.

![title](loop.png)

```python
i = 1
while i < 6:
    print(i)
    i += 2
```

```python
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1 
```

```python
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
```

```python
for x in "banana":
    print(x)
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
    if x == "banana":
        break
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        break
    print(x)
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        continue
    print(x)
```

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
    for y in fruits:
        print(x, y) 
```

```python
for x in range(6):
    print(x)
```

```python
for x in range(2, 6):
    print(x)
```
